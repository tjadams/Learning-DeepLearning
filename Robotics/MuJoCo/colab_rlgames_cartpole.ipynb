```python
cartpole_config = {'params': {'algo': {'name': 'a2c_discrete'},
  'config': {
   'clip_value': False,
   'critic_coef': 2,
   'e_clip': 0.2,
   'entropy_coef': 0.01,
   'env_config': {'env_name': 'CartPole-v1', 'seed': 5},
   'env_name': 'envpool',
   'full_experiment_name' : 'cartpole_onnx',
   'save_best_after' : 20,
   'gamma': 0.99,
   'grad_norm': 1.0,
   'horizon_length': 32,
   'kl_threshold': 0.008,
   'learning_rate': '3e-4',
   'lr_schedule': 'adaptive',
   'max_epochs': 100,
   'mini_epochs': 5,
   'minibatch_size': 1024,
   'name': 'cartpole',
   'normalize_advantage': True,
   'normalize_input': True,
   'normalize_value': True,
   'num_actors': 64,
   'player': {'render': True},
   'ppo': True,
   'reward_shaper': {'scale_value': 0.1},
   'schedule_type': 'standard',
   'score_to_win': 20000,
   'tau': 0.95,
   'truncate_grads': True,
   'use_smooth_clamp': False,
   'value_bootstrap': True},
  'model': {'name': 'discrete_a2c'},
  'network': {'mlp': {'activation': 'elu',
    'initializer': {'name': 'default'},
    'units': [32, 32]},
   'name': 'actor_critic',
   'separate': False,
   'space': {'discrete': {},
  'seed': 5}}
  }
}
```

```python
import yaml
from rl_games.torch_runner import Runner

config = cartpole_config
config['params']['config']['full_experiment_name'] = 'cartpole'
config['params']['config']['max_epochs'] = 500
config['params']['config']['horizon_length'] = 512
config['params']['config']['num_actors'] = 8
config['params']['config']['minibatch_size'] = 1024

runner = Runner()
runner.load(config)
runner.run({
    'train': True,
})
```
